```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

print('Shape:', df.shape)       # Should be (7043, 21)
print(df.head())
print(df.dtypes)
print(df['Churn'].value_counts())
```

```python
# TotalCharges has hidden missing values (blank strings)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print('Missing values:')
print(df.isnull().sum())
# You'll see 11 NaN rows in TotalCharges - we'll drop these
```

```python
fig = px.pie(
    df, names='Churn',
    title='Churn Distribution',
    color_discrete_sequence=['#636EFA', '#EF553B']
)
fig.show()
# ~26% churn — this is a class imbalance we'll handle in the model
```

```python
# Monthly Charges vs Churn
fig = px.box(
    df, x='Churn', y='MonthlyCharges',
    title='Monthly Charges by Churn Status',
    color='Churn'
)
fig.show()

# Contract Type vs Churn Rate
churn_by_contract = df.groupby('Contract')['Churn'].apply(
    lambda x: (x == 'Yes').mean() * 100
).reset_index()
churn_by_contract.columns = ['Contract', 'Churn Rate %']
print(churn_by_contract)
# Month-to-month has ~42% churn vs ~11% and ~3% for longer contracts
```
